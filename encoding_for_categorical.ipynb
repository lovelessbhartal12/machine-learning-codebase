```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/customer.csv')
df.head()
```

```python
# Extract label columns
y = df[['review', 'education', 'purchased']]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(y.iloc[:,0:2], y.iloc[:,2:], test_size=0.2, random_state=42)
```

```python
y_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
X_train = enc.fit_transform(X_train)
X_test = enc.transform(X_test)
```

```python
x_train = pd.DataFrame(X_train, columns=['review', 'education'])
x_test = pd.DataFrame(X_test, columns=['review', 'education'])
x_train.head()
```
