```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/datasets_11657_16098_train.xls')[['Age','Pclass','SibSp','Parch','Survived']]
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df['family_size']=df['SibSp']+df['Parch']+1
df.head()
```

```python
def myfun(num):
  if num==1:
    #small family
    return 0
  elif num>1 and num<4:
    #medium family
    return 1
  else:
    return 2
```

```python
myfun(5)
```

```python
df['family_type']=df['family_size'].apply(myfun)
df.head()
```

```python
df.drop(columns=['SibSp','Parch','family_size','Survived'])
```
